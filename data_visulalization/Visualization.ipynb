```python
# Density Plots
# The Density Plot is a variation of historgram that shows the
# smoothed distribution of the points along the numeric axis. 
# The peaks of the density plot are at locations where 
# there is high concentration of points.

hrdata.Income.plot('density',ls = '--')
plt.xlabel("Income")
plt.title("Density Plot for Income")
plt.rcParams['xtick.labelsize']=8
plt.tight_layout()
plt.show()
# We can give a dash line for the density plot by giving ls option
# which basically stands for the line style.
# Using the .xlabel method, we can modify the label for the x axis.
# Python has some resource configuration (rc) settings that are stored
# in a dictionary-like variable called matplotlib.rcParams, 
# You can use rcParams dictionary to change generic plot parameters 
# like xlabel tick size, font size, color scheme, etc .
# We can make the plot concise by using the tight layout method
# The show method of pyplot is used to display the plot.
```

```python
# Superimposing density plot over histogram
ax = hrdata['Income'].plot(kind = 'hist', color = 'lightblue',ylim=(0,400))
ax2 = hrdata['Income'].plot(kind = 'density', secondary_y = True, color = 'blue',ax=ax)
ax.set_ylabel('Number of Employees')
ax2.set_ylabel('Density')
# The histogram is plotted on the primary y axis.
# The ylim option helps in setting the minimum and maximum value
# on the primary y axis.
# The argument ax=ax,  sets the same axes objects for the density plot.
# The same axes gets activated to plot the density plot.
# By saying secondary_y=true, we are activating the secondary y axes
# to plot the density plot.
# We can set the y label by using the set_ylabel function.
```

```python
# Box plots
plt.boxplot(hrdata['Income'], vert=False)
# The default orientation of  boxplot is vertical. 
# If we want to change the orientation of boxplot,
# we can simply say vert=false.
# Besides the boxplot itself, we also get a dictionary. 
# The key fliers can tell us which values of the variable income
# lie in the extreme observations that skew the data.
```

```python
# To look at the actual extreme observations in the boxplot, 
# we save the boxplot in a variable.
bp = plt.boxplot(hrdata['Income'])
bp
```

```python
# Now we can access the key fliers from the output dictionary of bp 
# We then choose the first element and save this object as a,
a=bp['fliers'][0]
```

```python
# We can convert this into an array using the get_data option. 
# So here we get an array of all the extreme observations in 
# the column income from the HR data.  
a.get_data()[1]
```

```python
np.min(list(a.get_data()[1]))
# All above 16595 are outliers
```

```python
#Line charts are also very widely used in studying the overall 
#trend in the data. 
#When we use matplotlib, the default plot using the plt.plot function
#is a line plot, so if we say plt.plot variable name 
#what we get is a line plot.  

plt.plot(mmx['NewVolSales'], c = "M", ls = "--", marker = "s", ms=6 )
plt.xlabel("Week")
plt.ylabel("Volume Sales")
plt.title("Volume Sales of Cough Syrup")
plt.xticks(rotation = "vertical")
plt.show()

# c is a short form for colour and M stands for magenta. 
# ls is the line style 
# marker are the connecting points : s stands for square;
# ms is the marker size
# If we want ticks on x-axis to be vertical, we can do so by 
# using the plt.xticks function.
# We can label the x-axis using the x label method and 
# display the plot with the plt.show
```

```python
# Bivariate plots
# Scatter plots
plt.scatter(x=hrdata["Age"], y= hrdata["Income"], color = "Magenta", alpha=0.5)
plt.title("Relationship between Age and Income")
plt.xlabel("Age")
plt.ylabel("Income")
```

```python
# Another way of rendering Multiple Plots

# We can render multiple plots in the same plot area using 
# either Subplots or FacetGrid function 

# To plot the volume sales on the primary y axes and radio TRPs
# in the secondary y axes.

fig, ax1 = plt.subplots()

ax1.plot(mmx["NewVolSales"])
ax1.set_xlabel("Time in weeks")
ax1.set_ylabel("New Volume Sales")

ax2 = ax1.twinx()
ax2.plot(mmx["Radio "], 'r')
ax2.set_ylabel("Radio GRP")

fig.tight_layout()
plt.show()

# plt.subplots() returns 2 objects: figure and subplot. 
# Figure is the entire plot area in which there is a sub plot and
# within the sub plot, axes can be activated to render various plots.
# Axes contains figure elements including x label, y label, 
# title, type of the plot to be rendered, ticks etc.
# We use twinx option on the axes which is a part of the axes
# module to set the secondary y axes.
```

```python
# Using facet_grid
sb.FacetGrid(col='Department', aspect=1.6, size=4, col_wrap=5, data=hrdata).map(plt.scatter,\
                'Age', 'Income').fig.subplots_adjust(wspace=0.3,hspace=0.2)
# col_wrap “Wraps” the column variable after 5 departments
# Map is a method within seaborn which renders the plot that we would like.
# fig.subplots_adjust tunes the subplot layout.
# wspace = 0.3 is the amount of width reserved for space between subplots,
# expressed as a fraction of the average axis width
#hspace = 0.2 is the amount of height reserved for space between subplots,
# expressed as a fraction of the average axis height
```

```python
# We can customise this plot further by adding more options by 
# adjusting the subplots in the figure.
# Similarly we can plot the faceted charts on two different rows, 
# for example we would like to look at the distribution of income
# by gender for different marital status, this is how we can 
# specify the options within facetgrid and even arrange the
# order of the plots.  

g=sb.FacetGrid(col='MaritalStatus', aspect=1.6, size=4, data= hrdata,
    row='Gender', col_order= ['Single','Married','Divorced']).map(plt.hist,
     'Income', color = "m").fig.subplots_adjust(wspace=0.3,hspace=0.2)
```
